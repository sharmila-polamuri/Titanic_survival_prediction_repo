```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the training dataset
train_df = pd.read_csv('/Users/apple/Documents/Projects/titanic-survival-prediction/data/train.csv')

# Display the first few rows of the dataset
train_df.head()
```

```python
# Check for missing values
missing_values = train_df.isnull().sum()
print(missing_values)
```

```python
# Data Visualization
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Survived', palette='Set1')
plt.title('Survival Distribution (0: Not Survived, 1: Survived)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Pclass', palette='Set2')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(train_df['Age'].dropna(), kde=True, color='green')
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Sex', palette='Set3')
plt.title('Gender Distribution')
plt.show()
```

```python
# Visualize the relationship between the number of siblings/spouses and survival
sns.countplot(x='SibSp', hue='Survived', data=train_df)
plt.title('Survival by Number of Siblings/Spouses')
plt.show()
```

```python
# Visualize the relationship between the number of parents/children and survival
sns.countplot(x='Parch', hue='Survived', data=train_df)
plt.title('Survival by Number of Parents/Children')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Embarked')
plt.title('Port of Embarkation Distribution')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = train_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot
sns.pairplot(train_df, hue='Survived', palette='Set1')
plt.title('Pairplot')
plt.show()
```
